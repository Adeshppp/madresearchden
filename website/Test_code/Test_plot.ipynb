```python
np.random.seed(42)

temp1 = data.copy()

temp1['Cell'] = pd.Series([f"Cell {i}" for i in range(1,34)])
temp1['Cell Type'] = 'Cell Type 1'
temp1['Cluster X'] = np.random.normal(0.3, 0.05, 33)
temp1['Cluster Y'] = np.random.normal(0.4, 0.05, 33)
temp1['Gene Data'] = np.random.normal(1500, 700, 33)
temp1['Gene'] = ' Gene X'

temp2 = data.copy()

temp2['Cell'] = pd.Series([f"Cell {i}" for i in range(34,67)])
temp2['Cell Type'] = 'Cell Type 2'
temp2['Cluster X'] = np.random.normal(0.9, 0.05, 33)
temp2['Cluster Y'] = np.random.normal(0.2, 0.05, 33)
temp2['Gene Data'] = np.random.normal(500, 200, 33)
temp2['Gene'] = 'Gene Y'

temp2


temp3 = data.copy()

temp3['Cell'] = pd.Series([f"Cell {i}" for i in range(67,100)])
temp3['Cell Type'] = 'Cell Type 3'
temp3['Cluster X'] = np.random.normal(0.6, 0.05, 33)
temp3['Cluster Y'] = np.random.normal(0.5, 0.05, 33)
temp3['Gene Data'] = np.random.normal(1000, 500, 33)
temp3['Gene'] = 'Gene Z'

temp3




```

```python
np.random.seed(42)

data1_transform = np.random.normal(1, 0.5, 5)
data2_transform = np.random.normal(1, 0.1, 5)
data3_transform = np.random.normal(1.5, 0.5, 5)



```

```python
def make_data(df, transform, main_df):
    
    main_df = main_df.append(df)
    
    
    for i in range(0, len(transform)):
        
        temp = temp1.copy()
        temp['Gene Data'] = temp['Gene Data'] * transform[i]      
        main_df = main_df.append(temp)
    
    return main_df
    
    
        
        
data = make_data(temp1, data1_transform, data)
data = make_data(temp2, data2_transform, data)
data = make_data(temp3, data3_transform, data)

data
# make_data(temp2)
# make_data(temp3)
        
        
```
